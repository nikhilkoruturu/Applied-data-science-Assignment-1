```python
import pandas as pd

# Replace 'your_dataset.csv' with the path to your CSV file
file_path = '/content/US Stock Market Dataset.csv'

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify it loaded correctly
print(df.head(5))
```

```python
# Display basic information about the DataFrame
print("Data Info Before Cleaning:")
print(df.info())
```

```python
# Handle missing values
# Example: Replace missing numerical values with the mean of the column
df.fillna(df.mean(), inplace=True)
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Convert data types if needed
# Example: Convert a column to datetime
# df['date_column'] = pd.to_datetime(df['date_column'])

# Perform any other necessary cleaning operations

# Display basic information about the cleaned DataFrame
print("\nData Info After Cleaning:")
print(df.info())
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Compute the correlation matrix
correlation_matrix = df.corr()

# Print correlation matrix
print("n\Correlation Matrix:")
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filter the DataFrame for the range of years from 2020 to 2024
df_filtered = df[df['Date'].dt.year.between(2020, 2024)]

# Group by year and aggregate the data (e.g., taking the mean or sum)
df_aggregated = df_filtered.groupby(df_filtered['Date'].dt.year).mean()  # You can also use sum() instead of mean()

# Plot bar graph using matplotlib
plt.figure(figsize=(8, 6))
ax = df_aggregated.plot(kind='bar', ax=plt.gca())
plt.title('Bar Graph for the Years 2020 to 2024')
plt.xlabel('Year')
plt.ylabel('Mean Value')  # You can adjust ylabel according to your data
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
# Plotting the line graph
plt.figure(figsize=(8, 6))

# Assuming 'Date' column is your x-axis and 'Natural_Gas_Price' is one of the y-axes
plt.plot(df['Date'], df['Natural_Gas_Price'], label='Natural Gas Price')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Natural Gas Price Over Time')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add grid for better visualization
plt.grid(True)

# Add legend
plt.legend()

# Show plot
plt.show()
```
